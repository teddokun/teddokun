# **DATA CLEANING Phase 3: Column-Specific Parsing & Smart Imputation**

```python
# Instead of removing the missing value of salary_estimate we created three brand new columns (min_salary, max_salary, and avg_salary). We applied the mean/median imputation only to those three new numerical columns.
```

```python
salary_cleaned = df['salary_estimate'].str.split('(').str[0]
salary_cleaned = salary_cleaned.str.replace('K', '').str.replace('$', '').str.strip()
```

```python
min_str = salary_cleaned.str.split('-').str[0]
max_str = salary_cleaned.str.split('-').str[1]
```

```python
df['min_salary'] = pd.to_numeric(min_str, errors='coerce') * 1000
df['max_salary'] = pd.to_numeric(max_str, errors='coerce') * 1000
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2
```

```python
# Check Skewness and Impute Missing Values
salary_skewness = df['avg_salary'].skew()
print(f"Skewness of Average Salary: {salary_skewness:.2f}")

if abs(salary_skewness) > 0.5:
    print("-> Distribution is skewed. Imputing missing values with the MEDIAN.")
    fill_min = df['min_salary'].median()
    fill_max = df['max_salary'].median()
    fill_avg = df['avg_salary'].median()
else:
    print("-> Distribution is relatively symmetrical. Imputing missing values with the MEAN.")
    fill_min = df['min_salary'].mean()
    fill_max = df['max_salary'].mean()
    fill_avg = df['avg_salary'].mean()
```

```python
df['min_salary'].fillna(fill_min, inplace=True)
df['max_salary'].fillna(fill_max, inplace=True)
df['avg_salary'].fillna(fill_avg, inplace=True)
```

```python
df['company_name'] = df['company_name'].astype(str).str.split('\n').str[0]
df['company_name'] = df['company_name'].replace('nan', np.nan)
```

```python
import re

def parse_company_size(size_str):
    if pd.isna(size_str):
        return np.nan, np.nan

    size_str = size_str.replace(' employees', '').strip()

    if 'to' in size_str:
        parts = size_str.split(' to ')
        min_size = int(parts[0])
        max_size = int(parts[1])
    elif '+' in size_str:
        min_size = int(size_str.replace('+', ''))
        max_size = np.inf
    else:
        return np.nan, np.nan

    return min_size, max_size

df[['min_employees', 'max_employees']] = df['size'].apply(lambda x: pd.Series(parse_company_size(x)))

df[['size', 'min_employees', 'max_employees']].head(10)
```

# **DATA CLEANING Phase 4: Feature Engineering from 'Job Description'**

```python
import re
import numpy as np

def extract_experience_strict(text):
    if pd.isna(text):
        return np.nan

    text = str(text).lower()

    word_to_num = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5,
                   'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10}

    pattern = re.compile(r'((?:\d+|one|two|three|four|five|six|seven|eight|nine|ten))\s*(?:\+|plus|or more)?(?:\s*-\s*\d+)?\s*years?(?:\s*of)?(?:[a-z\s-]{0,30})?experience')

    matches = list(pattern.finditer(text))

    company_brag_words = ['we ', 'our ', 'company', 'firm', 'history', 'success', 'past',
                          'business', 'provider', 'founded', 'team', 'combined', 'collective']

    valid_numbers = []

    for match in matches:
        start_index = max(0, match.start() - 60)
        context_window = text[start_index:match.start()]

        is_company_brag = any(flag in context_window for flag in company_brag_words)

        if not is_company_brag:
            val = match.group(1)

            if val in word_to_num:
                num = word_to_num[val]
            else:
                try:
                    num = float(val)
                except ValueError:
                    continue

            if num <= 15:
                valid_numbers.append(num)

    if valid_numbers:
        return min(valid_numbers)

    return np.nan

df['clean_yrs_exp'] = df['job_description'].apply(extract_experience_strict)

print("Updated Count of Specific Years of Experience Needed:")
display(df['clean_yrs_exp'].value_counts().sort_index())
```

```python
exp_bins = [-2, 0, 2, 5, 8, np.inf]
exp_labels = ['Not Specified/0', 'Junior (1-2 Yrs)', 'Mid-Level (3-5 Yrs)', 'Senior (6-8 Yrs)', 'Lead (9+ Yrs)']
df['job_level'] = pd.cut(df['clean_yrs_exp'].fillna(-1), bins=exp_bins, labels=exp_labels)

salary_bins = [0, 30000, 50000, 75000, 100000, 125000, 150000, 175000, 200000, np.inf]
salary_labels = ['<$30K', '$30K-$50K', '$50K-$75K', '$75K-$100K', '$100K-$125K', '$125K-$150K', '$150K-$175K', '$175K-$200K', '>$200K']
df['salary_range'] = pd.cut(df['avg_salary'], bins=salary_bins, labels=salary_labels, right=False)

print("--- Job Leveling Matrix: Salary vs. Required Experience ---")
leveling_matrix = pd.crosstab(df['salary_range'], df['job_level'])
display(leveling_matrix)
```

```python
def is_entry_level(title, description):
    title = str(title).lower()
    desc = str(description).lower()

    junior_titles = ['junior', 'jr', 'entry level', 'entry-level', 'graduate']
    junior_exp = ['0-1 year', '0-2 year', '1-2 year', 'fresh graduate', 'no experience']

    if any(word in title for word in junior_titles) or any(phrase in desc for phrase in junior_exp):
        return True
    return False

df['is_entry_level'] = df.apply(lambda row: is_entry_level(row['job_title'], row['job_description']), axis=1)

df_entry = df[df['is_entry_level'] == True]
print(f"Total Entry-Level Jobs found: {len(df_entry)}")
```

```python
hard_skills_list = [
    "sql", "python", "excel", "tableau", "power bi", "machine learning",
    "data visualization", "predictive analytics", "data mining", "data modeling",
    "statistics", "data warehousing", "ux/ui design", "agile", "scrum",
    "data governance", "business intelligence", "data engineering", "etl"
]

soft_skills_list = [
    "communication", "problem solving", "teamwork", "critical thinking",
    "curiosity", "project management", "data storytelling", "business acumen",
    "leadership", "collaboration", "time management", "attention to detail"
]

descriptions = df['job_description'].astype(str).str.lower()

hard_counts = {skill.title(): descriptions.str.contains(skill, na=False).sum() for skill in hard_skills_list}
soft_counts = {skill.title(): descriptions.str.contains(skill, na=False).sum() for skill in soft_skills_list}

df_hard = pd.DataFrame(list(hard_counts.items()), columns=['Skill', 'Frequency']).sort_values(by='Frequency', ascending=False)
df_soft = pd.DataFrame(list(soft_counts.items()), columns=['Skill', 'Frequency']).sort_values(by='Frequency', ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

sns.barplot(x='Frequency', y='Skill', data=df_hard.head(10), ax=axes[0], palette='Blues_r')
axes[0].set_title('Top 10 Hard Skills for Data Analysts', fontsize=16, fontweight='bold')
axes[0].set_xlabel('Number of Job Postings', fontsize=12)
axes[0].set_ylabel('')

sns.barplot(x='Frequency', y='Skill', data=df_soft.head(10), ax=axes[1], palette='Greens_r')
axes[1].set_title('Top 10 Soft Skills for Data Analysts', fontsize=16, fontweight='bold')
axes[1].set_xlabel('Number of Job Postings', fontsize=12)
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
df_entry.loc[:, 'requires_degree'] = df_entry['job_description'].str.contains(r"bachelor|degree|ba/bs", case=False, na=False)
df_entry.loc[:, 'accepts_equivalent'] = df_entry['job_description'].str.contains(r"equivalent experience|bootcamp|or equivalent", case=False, na=False)

degree_req_pct = (df_entry['requires_degree'].sum() / len(df_entry)) * 100
equiv_acc_pct = (df_entry['accepts_equivalent'].sum() / len(df_entry)) * 100

print(f"Jobs requiring a degree: {degree_req_pct:.1f}%")
print(f"Jobs explicitly accepting equivalent experience (great for shifters!): {equiv_acc_pct:.1f}%")

req_data = {'Requirement': ['Requires Degree', 'Accepts Equivalent'],
            'Percentage': [degree_req_pct, equiv_acc_pct]}
req_df = pd.DataFrame(req_data)

custom_palette = []
for requirement in req_df['Requirement']:
    if requirement == 'Requires Degree':
        custom_palette.append('#3764C5')
    elif requirement == 'Accepts Equivalent':
        custom_palette.append('#4AAD44')

plt.figure(figsize=(8, 6))
sns.barplot(data=req_df, x='Requirement', y='Percentage', hue='Requirement', palette=custom_palette, legend=False)
plt.title('The Education Barrier: Degree vs. Equivalent Experience', fontsize=14)
plt.ylabel('Percentage of Entry-Level Postings (%)', fontsize=12)
plt.ylim(0, 100)

plt.grid(False)

for index, row in req_df.iterrows():
    plt.text(index, row.Percentage + 2, f"{row.Percentage:.1f}%", color='black', ha="center", fontsize=12, fontweight='bold')
plt.show()
```

# **DATA CLEANING Phase 5: Final Validation**

```python
import pandas as pd
import numpy as np

if 'salary_estimate' in df.columns:
    df.drop('salary_estimate', axis=1, inplace=True)

if 'headquarters' in df.columns:
    df['hq_city'] = df['headquarters'].astype(str).str.split(',').str[0].str.strip()
    df['hq_state'] = df['headquarters'].astype(str).str.split(',').str[1].str.strip()
    df.drop('headquarters', axis=1, inplace=True)

if 'founded' in df.columns:
    df['company_age'] = 2026 - df['founded']
    df.drop('founded', axis=1, inplace=True)

df['rating'] = df['rating'].fillna(df['rating'].median())
df['company_age'] = df['company_age'].fillna(df['company_age'].median())

text_cols_to_fill = ['size', 'type_of_ownership', 'industry', 'sector',
                     'revenue', 'competitors', 'easy_apply', 'hq_city', 'hq_state']
for col in text_cols_to_fill:
    if col in df.columns:
        df[col] = df[col].fillna('Not Specified')

int_cols = ['min_salary', 'max_salary', 'avg_salary', 'company_age']
for col in int_cols:
    if col in df.columns:
        df[col] = df[col].astype(int)

category_cols = ['size', 'type_of_ownership', 'industry', 'sector',
                 'revenue', 'hq_state', 'job_state']

for col in category_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

string_cols = ['job_title', 'job_description', 'company_name',
               'hq_city', 'job_city', 'competitors']

for col in string_cols:
    if col in df.columns:
        df[col] = df[col].astype('string')

print("--- Final Dataset Information ---")
print(df.info())
```
